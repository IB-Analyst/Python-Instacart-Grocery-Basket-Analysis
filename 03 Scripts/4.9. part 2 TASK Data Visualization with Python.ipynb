### Explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.

```python
# Reducing previously made sample to only the columns needed

custs_ords_prods_merged_4 = small[['age','income']]
```

```python
custs_ords_prods_merged_4.head()
```

```python
# Creating a scatterplot

scatterplot = sns.scatterplot(x = 'age', y = 'income',data = custs_ords_prods_merged_4)
```

##### Customers under the age of 40 mostly have an income up to 200,000 and some - up to 400,000. 
##### Consumers aged 40 and over mostly have an income up to 300,000 and some - up to 600,000. That is, we can conclude that, after all, there is a connection between age and income. 
##### The turning point in the difference is 40 yearsold customers.Can say that the older the buyer, the more likely he has a higher income.

```python
# Exporting line chart

scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatterplot_age_income_task4.9.png'))
```
